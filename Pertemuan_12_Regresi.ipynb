```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
usia_mobil = [5, 4, 6, 5, 5, 5, 6, 6, 2, 7, 7]
harga_mobil = [85, 103, 70, 82, 89, 98, 66, 95, 169, 70, 48]
```

```python
X = np.array(usia_mobil).reshape(-1, 1)
y = np.array(harga_mobil)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
slope = model.coef_[0]  # m in y = mx + b
intercept = model.intercept_  # b in y = mx + b
```

```python
print(f"Persamaan regresi: y = {slope:.2f}x + {intercept:.2f}")
```

```python
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, model.predict(X), color='red', label='Regresi')
plt.title("Regresi Linear Harga Mobil vs. Usia Mobil")
plt.xlabel("Usia Mobil (tahun)")
plt.ylabel("Harga Mobil ($100)")
plt.legend()
plt.show()
```
